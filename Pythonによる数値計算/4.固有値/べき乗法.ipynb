## べき乗法

行列Aの最大固有値を出力する

### べき乗法のアルゴリズム

１）A,x(0)を入力する

２）k=0,1,2,....に対して

　　
  a)y(k)=Ax(k),　　λ=(y(k),y(k))/(y(k),x(k))
  
 　 　b)x(k+1)=y(k)/|y(k)|
 
 3)λの値が収束すれば反復を終了する

```python
import math

NM=3
y=[0]*NM
EPS=1.0E-6

#1
n=3
a=[[11,7,-5],[0,10,-1],[2,8,3]]

x=[1.0]*n

#2
kmax=200
lambda1 = 10000
lambda2 = 10000
for k in range(1,kmax+1):
    gosa=0
    #(a) y=Ax
    for i in range(n):
        y[i]=0
        for j in range(n):
            y[i]+=a[i][j]*x[j]
            
    #(a) λ=(y,y)/(y,x)
    sum1 = 0
    sum2 = 0
    for i in range(n):
        sum1+=y[i]*y[i]
        sum2+=y[i]*x[i]
    lambda1 = sum1/sum2
    
    #(b)
    for i in range(n):
        x[i]=y[i]/math.sqrt(sum1)
    
    #3
    gosa=abs(lambda1-lambda2)/abs(lambda2)
    lambda2=lambda1
    if gosa < EPS:
        print("Converged!")
        print()
        break
        
    print(f'k={k:3} lambda={lambda1:10.6f}')
    
if k >= kmax:
    print()
    print("It didn't converged.")
```

pythonのモジュールによる出力

```python
import numpy as np
import numpy.linalg as LA
A = np.array([[11,7,-5],
             [0,10,-1],
             [2,8,3]])

w,v=LA.eig(A)
print(w)
print(v)
```

出力値wは行列Aの固有値となる

最大固有値が９であることがわかる

逆に最小固有値は７であることも同時にわかる
