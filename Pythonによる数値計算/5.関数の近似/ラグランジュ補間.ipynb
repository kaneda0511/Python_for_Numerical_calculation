```python
# ラグランジュ補間
import numpy as np

# Pnを関数として定義
def Pn(XX,X):
    for i in range(0,N):
        numer=1.0 #分子
        denom=1.0 #分母
        for j in range(0,N):
            if j!=i:
                numer=numer*(XX-X[j])
                denom=denom*(X[i]-X[j])
            L[i]=numer/denom
        
    P=0.0
    for i in range(0,N):
        P=P+Y[i]*L[i]
    return P
```
