```python
import math

def f(t,x):
    return (t**2 + t + 1.0) - (2.0*t + 1.0) * x + x ** 2

def exact(t):
    return (t*math.exp(t)+t+1.0)/(math.exp(t)+1.0)

NM = 20
t = [0]*(NM+1)
x = [0]*(NM+1)
m = 20
h = 0.1
t[0] = 0.0
x[0] = 0.5 #初期条件
q0 = 0.0

for n in range(m):
    t[n+1] = t[n] + h

    k0 = h*f(t[n],x[n])
    r1 = (k0-2*q0)/2
    x[n+1] = x[n] + r1
    q1 = q0 + 3 * r1 -k0/2

    k1 = h * f(t[n]+h/2,x[n+1])
    r2 = (1 - 1/math.sqrt(2)) * (k1-q1)
    x[n+1] = x[n+1] + r2
    q2 = q1 + 3 *r2 - (1-1/math.sqrt(2))*k1

    k2 = h * f(t[n]+h/2,x[n+1])
    r3 = (1+1/math.sqrt(2))*(k2-q2)
    x[n+1] = x[n+1] + r3
    q3 = q2 + 3 * r3 - (1+1/math.sqrt(2))*k2

    k3 = h * f(t[n]+h,x[n+1])
    r4 = (k3 - 2*q3) / 6
    x[n+1] = x[n+1] + r4
    q4 = q3 + 3 * r4 -k3 / 2

    q0 = q4

print('{:^6}{:^15}{:^15}'.format('t(n)','calculated','exact'))
for n in range(m+1):
    print(f'{t[n]:6.2f}{x[n]:15.6f}{exact(t[n]):15.6f}')
```

```python
print('{:^6}{:^15}{:^15}{:^15}'.format('t(n)','calculated','exact','diff'))
for n in range(m+1):
    print(f'{t[n]:6.2f}{x[n]:15.6f}{exact(t[n]):15.6f}{x[n]-exact(t[n]):15.6f}')
```
