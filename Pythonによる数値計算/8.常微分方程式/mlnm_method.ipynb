```python
import math

def f(t,x):
    return (t**2 + t + 1.0) - (2.0*t + 1.0) * x + x ** 2

def exact(t):
    return (t*math.exp(t)+t+1.0)/(math.exp(t)+1.0)

NM = 20
t = [0]*(NM+1)
x = [0]*(NM+1)
g = [0]*(NM+1)
m = 20
EPS=1e-77
h = 0.1
t[0] = 0.0
x[0] = 0.5 #初期条件

#ルンゲ・クッタ法
for n in range(m):
    t[n+1] = t[n] + h
    k1 = h*f(t[n],x[n])
    k2 = h*f(t[n]+h/2,x[n]+k1/2)
    k3 = h*f(t[n]+h/2,x[n]+k2/2)
    k4 = h*f(t[n]+h,x[n]+k3)
    x[n+1] = x[n] + (k1+ 2*k2 + 2*k3 + k4)/6


f1 = f(t[1],x[1])
f2 = f(t[2],x[2])
f3 = f(t[3],x[3])

n = 3
g[n+1] = x[n-3] + 4 * h * (2*f1 - f2 + 2*f3) / 3

for n in range(3,m):
    while True:
        ff = f(t[n+1],g[n+1])

        x[n+1] = x[n-1] + h * (f(t[n-1],x[n-1]) + 4 * f(t[n],x[n]) + ff) / 3

        if abs(g[n+1]-x[n+1])<EPS:
            break
        else:
            g[n+1]=x[n+1]

        t[n+1] = t[n] + h


print('{:^6}{:^15}{:^15}'.format('t(n)','calculated','exact'))
for n in range(m+1):
    print(f'{t[n]:6.2f}{x[n]:15.6f}{exact(t[n]):15.6f}')
```
