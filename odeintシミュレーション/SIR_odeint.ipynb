```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def SIR_EQ(v,t,beta,gamma):
    f1 = -beta*v[0]*v[1]
    f2 = beta*v[0]*v[1] - gamma*v[1]
    f3 = gamma*v[1]
    return [f1,f2,f3]
```

```python
t = np.arange(0,10,0.01)
S_0,I_0,R_0 = map(int,input().split())
N0 = [S_0,I_0,R_0]
args=(0.0026,0.5)
result = odeint(SIR_EQ,N0,t,args)

plt.plot(t,result)
```
