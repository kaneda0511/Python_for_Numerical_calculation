```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def Lotka_Volterra(v,t,a,b,c,d):
    dxdt = a*v[0] - b*v[0]*v[1]
    dydt = c*v[0]*v[1] - d*v[1]
    return [dxdt,dydt]
```

```python
a,b,c,d = 1,0.4,0.2,1
t = np.arange(0,40,0.01)
args = (a,b,c,d)
x0 = 10
y0 = 2
init_stats = (x0,y0)
result = odeint(Lotka_Volterra,init_stats,t,args)
plt.plot(t,result)
plt.legend(["x","y"])
plt.show()
```

```python
def Lotka_Volterra2(v,t,a,b,c,d):
    if 0< t < 20:
        dxdt = a*v[0] - b*v[0]*v[1]
        dydt = c*v[0]*v[1] - d*v[1]
        return [dxdt,dydt]
    else:
        dxdt = a*v[0] - 2*b*v[0]*v[1]
        dydt = c*v[0]*v[1] - 2*d*v[1]
        return [dxdt,dydt]
```

```python
a,b,c,d = 1,0.4,0.2,1
t = np.arange(0,40,0.01)
args = (a,b,c,d)
x0 = 10
y0 = 2
init_stats = (x0,y0)
result = odeint(Lotka_Volterra2,init_stats,t,args)
plt.plot(t,result)
plt.legend(["x","y"])
plt.show()
```
