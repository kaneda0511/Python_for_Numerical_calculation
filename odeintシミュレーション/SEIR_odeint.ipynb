```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def SEIR(v,t,beta,ip,gamma):
    N = 763
    fs = -beta*v[0]*v[2]/N
    fe = beta*v[0]*v[2]/N - (1/ip)*v[1]
    fi = (1/ip)*v[1] - (1/gamma)*v[2]
    fr = (1/gamma)*v[2]
    return [fs,fe,fi,fr]
```

```python
N,S0,E0,I0 = 762,0,1,0
init_state = [N,S0,E0,I0]
t = np.arange(0,14,0.01)
beta,lp,gamma=6.87636378, 1.21965986, 2.01373496
args = (beta,ip,gamma)
result = odeint(SEIR,init_state,t,args)
```

```python
plt.plot(t,result)
plt.legend(['Susceptible','Exposed','Infected','Recovered'])
```
