# 1階微分方程式
### http://quantum.phys.tohoku.ac.jp/odeint

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def dxdt(x,t):
    dxdt = -x
    return dxdt
```

```python
t = np.arange(0,10,0.1)
x0 = [0,1]
```

```python
x = odeint(dxdt,x0,t)
plt.plot(t,x[:,1])
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

# 練習問題

```python
# 3.1.1
def dxdt(x,t,a,b):
    dxdt = -(b+x**2)/(a*t**2)
    return dxdt

#　初期条件x(1)=2
x0 = [2,1]
t = np.arange(1,10,0.1)

# args=(定数a,定数b)
x = odeint(dxdt,x0,t,args=(2,5))
#　初期条件を満たすのはx[:,1]
plt.plot(t,x[:,1])
plt.show()
```

```python
# 3.1.2
def dNdt(N,t):
    dNdt = -r*N
    return dNdt

N0 = [10000,0]
t = np.arange(0,20000,1)
r = 1.2096e-4
ans = odeint(dNdt,N0,t)
plt.plot(t,ans[:,0])
plt.show()
```
