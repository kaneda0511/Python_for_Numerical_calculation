## 自由落下のシミュレーション

```python
#　重力加速度
G = 9.80665

#　時刻
t = 0
#　時刻の刻み幅
h = 0.01

#　初速度v0の入力
v = float(input())

#　初期位置x0の入力
x = float(input())

#　現在時刻と現在の位置
print("{:.7f} {:.7f}".format(t, x, v))

#　自由落下の計算
while x >= 0:     #　地面に達するまで計算
    t += h        #　時刻の更新
    v += G*h      #　速度の計算
    x -= v*h      #　位置の更新
    print("{:.7f} {:.7f} {:.7f}".format(t, x, v))

```
