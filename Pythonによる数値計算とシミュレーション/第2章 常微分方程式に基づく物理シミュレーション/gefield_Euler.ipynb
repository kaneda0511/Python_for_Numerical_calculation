# 二次元運動のシミュレーション
## 電界中の電化粒子のシミュレーション

```python
import math
import matplotlib.pyplot as plt
import numpy as np

#　電荷の位置と値
Q = (((0.0,0.0),10.0),((5.0,-5.0),5.0))

TIMELIMIT = 20.0         #　シミュレーション打ち切り時刻
RLIMIT = 0.1             #　距離rの最低値
H = 0.01                 #　時刻の刻み幅
t = 0.0                  #　時刻t

#　電荷位置のプロット
for qi in Q:
    plt.plot(qi[0][0],qi[0][1],".")

vx = float(input())     #　初速度v0x
vy = float(input())     #　初速度v0y
x = float(input())      #　初期位置x
y = float(input())      #　初期位置y

print("{:.7f} {:.7f} {:.7f} {:.7f} {:.7f}".format(t,x,y,vx,vy))

#　グラフデータに現在位置を追加
xlist = [x]
ylist = [y]

#　二次元運動の計算
while t < TIMELIMIT:                 #　打ち切り時間まで計算
    t += H                           #　時刻の更新
    rmin = float("inf")              #　距離の最小化を初期化
    for qi in Q:
        rx = qi[0][0] - x               #　距離rxの計算
        ry = qi[0][1] - y               #　距離ryの計算
        r = math.sqrt(rx*rx + ry*ry)    #　距離rの計算
        if r < rmin:
            rmin = r                   #　距離の最小値を更新                
        vx += (rx/r/r/r*qi[1]) * H     #　速度vxの計算
        vy += (ry/r/r/r*qi[1]) * H     #　速度vyの計算
    x += vx * H     #　位置xの計算
    y += vy * H     #　位置yの計算
    print("{:.7f} {:.7f} {:.7f} {:.7f} {:.7f}".format(t,x,y,vx,vy))
    #　グラフデータに現在位置を追加
    xlist.append(x)
    ylist.append(y)
    if rmin < RLIMIT:
        break

#　グラフ表示
plt.plot(xlist,ylist)
plt.show()


```
