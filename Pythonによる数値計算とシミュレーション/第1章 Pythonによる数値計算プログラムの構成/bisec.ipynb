## 二分法による方程式の解法プログラム

#### 平方根の計算

```python
# グローバル変数
a = int(input())

#　終了条件
LIMIT = 1e-20

#　関数の定義
def f(x):
    return x*x -  a

#　初期設定
xp = float(input())
xn = float(input())

#　繰り返し処理
while (xp-xn) **2 > LIMIT:               #　終了条件を満たすまで繰り返す
    xmid = (xp+xn)/2                     #　新しい中間値の計算
    if f(xmid) > 0:                      #　中間値が正ならxpを更新　ここで定義した関数に中間点を代入している
        xp = xmid
    else:
        xn = xmid                        #　xnを更新
    print("{:15f} {:.15f}".format(xn,xp))
```

### 平方根の出力なら下記のコードの方がpythonライブラリーを使って簡単に求めることができる

```python
import math
a = int(input())
math.sqrt(a)
```

#### 立方棍の計算

```python
# グローバル変数
a = int(input())

#　終了条件
LIMIT = 1e-20

#　関数の定義
def f(x):
    return x**3 - a

#　初期設定
xp = float(input())
xn = float(input())

#　繰り返し処理
while (xp-xn) **3 > LIMIT:               #　終了条件を満たすまで繰り返す
    xmid = (xp+xn)/2                     #　新しい中間値の計算
    if f(xmid) > 0:                      #　中間値が正ならxpを更新　ここで定義した関数に中間点を代入している
        xp = xmid
    else:
        xn = xmid                        #　xnを更新
    print("{:15f} {:.15f}".format(xn,xp))
```
